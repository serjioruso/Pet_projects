<h3>1.1. Functions</h3>

```python
#Functions used in project

##Calculate mean target per category

def calculate_mean_target_per_category (data, variable):
    total_number_of_labels = len(data)
    
    temp_df = pd.Series(data[variable].value_counts() / total_number_of_labels).reset_index()
    temp_df.columns = [variable, 'perc_of_total']
    temp_df = temp_df.merge(
        data.groupby([variable])['SalePrice'].mean().reset_index(),
        on = variable,
        how = 'left'
    )
    
    return temp_df

##Find normal boundaries

def find_normal_boundaries (data, variable):
    lower_limit = data[variable].mean() - 3*data[variable].std()
    upper_limit = data[variable].mean() + 3*data[variable].std()
    
    return lower_limit, upper_limit

##Find skewed boundaries

def find_skewed_boundaries(data, variable, distance):
    IQR = data[variable].quantile(0.75) - data[variable].quantile(0.25)
    lower_boundary = data[variable].quantile(0.25) - (IQR*distance)
    upper_boundary = data[variable].quantile(0.75) + (IQR*distance)
    
    return lower_boundary, upper_boundary
```

<h3> Decision Tree Regression </h3>

```python
#Create pipeline
DTR_pipe = Pipeline(steps = [
    ('numerical_imputer', MeanMedianImputer(imputation_method='mean', variables = ['MasVnrArea'])),
    ('categorical_imputer', CategoricalImputer(imputation_method='frequent')),
    ('encoder', ce.CountEncoder()),
    ('scaler',RobustScaler(with_centering = False, with_scaling = True, quantile_range = (0, 100))),
    ('regressor', DecisionTreeRegressor(random_state=0))
])
    


X_train = X_train_copy.copy()
X_test = X_test_copy.copy()

#Fit pipeline
DTR_pipe.fit(X_train, y_train)


#GridSearch
parameters = [
    {
     'regressor__criterion' : ["squared_error", "friedman_mse", "absolute_error","poisson"],
     'regressor__splitter' : ["best","random"],
     'regressor__max_features' : ["auto", "sqrt", "log2"]
     }
]

grid_search = GridSearchCV(
    estimator = DTR_pipe,
    param_grid = parameters,
    scoring = 'accuracy',
    cv = 10,
    n_jobs = -1
)

grid_search.fit(X_train, y_train)
best_parameters = grid_search.best_params_
print('Best parameters: ', best_parameters)

model = grid_search.best_estimator_
```

```python
#Repeat pipeline with best parameters
DTR_pipe = Pipeline(steps = [
    ('numerical_imputer', MeanMedianImputer(imputation_method='mean', variables = ['MasVnrArea'])),
    ('categorical_imputer', CategoricalImputer(imputation_method='frequent')),
    ('encoder', ce.CountEncoder()),
    ('scaler',RobustScaler(with_centering = False, with_scaling = True, quantile_range = (0, 100))),
    ('regressor', DecisionTreeRegressor(random_state=0, criterion = 'absolute_error', max_features = 'auto', splitter = 'best'))
])

X_train = X_train_copy.copy()
X_test = X_test_copy.copy()

#Fit pipeline Train
DTR_pipe.fit(X_train, y_train)

#Prediction
X_train_preds = DTR_pipe.predict(X_train)
X_test_preds = DTR_pipe.predict(X_test)


train_MSE = mean_squared_error(y_train, X_train_preds)
train_RMSE = sqrt(mean_squared_error(y_train, X_train_preds))
train_r2 = r2_score(y_train, X_train_preds)
test_MSE = mean_squared_error(y_test, X_test_preds)
test_RMSE = sqrt(mean_squared_error(y_test, X_test_preds))
test_r2 = r2_score(y_test, X_test_preds)
                 
model_list.append('Decision Tree Regression')
train_MSE_list.append(train_MSE)
train_RMSE_list.append(train_RMSE)
train_r2_list.append(train_r2)
test_MSE_list.append(test_MSE)
test_RMSE_list.append(test_RMSE)
test_r2_list.append(test_r2)

print('Train MSE: {}'.format(mean_squared_error(y_train, X_train_preds)))
print('Train RMSE: {}'.format(sqrt(mean_squared_error(y_train, X_train_preds))))
print('Train r2: {}'.format(r2_score(y_train, X_train_preds)))
print()
print('Test MSE: {}'.format(mean_squared_error(y_test, X_test_preds)))
print('Test RMSE: {}'.format(sqrt(mean_squared_error(y_test, X_test_preds))))
print('Test r2: {}'.format(r2_score(y_test, X_test_preds)))
```

<h3>Random Forest Regression</h3>

```python
#Create pipeline
SVR_pipe = Pipeline(steps = [
    ('numerical_imputer', MeanMedianImputer(imputation_method='mean', variables = ['MasVnrArea'])),
    ('categorical_imputer', CategoricalImputer(imputation_method='frequent')),
    ('encoder', ce.CountEncoder()),
    ('scaler',RobustScaler(with_centering = False, with_scaling = True, quantile_range = (0, 100))),
    ('regressor', SVR(kernel = 'rbf'))
])
    


X_train = X_train_copy.copy()
X_test = X_test_copy.copy()

#Fit pipeline
SVR_pipe.fit(X_train, y_train)


#GridSearch
parameters = [
    {
     'regressor__kernel' : ['linear', 'poly', 'rbf', 'sigmoid', 'precomputed'],
     'regressor__gamma' : ['scale', 'auto']
     }
]

grid_search = GridSearchCV(
    estimator = SVR_pipe,
    param_grid = parameters,
    scoring = 'accuracy',
    cv = 10,
    n_jobs = -1
)

grid_search.fit(X_train, y_train)
best_parameters = grid_search.best_params_
print('Best parameters: ', best_parameters)

model = grid_search.best_estimator_
```
